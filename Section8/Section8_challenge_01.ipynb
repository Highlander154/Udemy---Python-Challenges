> * Write a Python program that prints if a number num is either "Positive", "Negative", or "Zero".

ðŸ”¹ Expected Output:



---
#### Function
---

```python
def positive_or_negative(num: int) -> str:
    if num > 0:
        return "Positive"
    if num < 0:
        return "Negative"
    return "Zero"
```

---
#### Tests
---

```python
print(positive_or_negative(3))
print(positive_or_negative(-1))
print(positive_or_negative(0))
```
