> * Write a Python program that prints the largest value in a dictionary.
> * If the dictionary is empty, print None.
> * You may assume that the values are numeric.

ðŸ”¹ Expected Output:



You may print "None" as a string or as the special value None.

```python
def largest_in_dict(dict_: dict) -> int|None:
    if dict_:
        return max(dict_.values())
    return None
```

```python
d = {"a": 4, "b": 3, "c": 7}
print(largest_in_dict(d))
```

```python
d = {"a": 4, "b": 6}
print(largest_in_dict(d))
```

```python
d = {"a": 4, "b": 4}
print(largest_in_dict(d))
```

```python
d = {}
print(largest_in_dict(d))
```
