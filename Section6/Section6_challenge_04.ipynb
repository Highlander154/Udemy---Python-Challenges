> * Write a Python program that checks if all values in a dictionary are equal.
> * If they are, print True. Else, print False.
> * If the dictionary is empty, print "Empty".

ðŸ”¹ Expected Output:



ðŸ”¸ Hints:
> * Loops will be very helpful to solve this challenge. You could also make a set from the dictionary to keep only one copy per value.
> * The .values() method returns the values in a dictionary.

```python
def all_equal(dict_: dict) -> bool:
    if not dict_:
        return "Empty"
    elif len(set(dict_.values())) == 1:
        return True
    return False
```

```python
d = {"a": 4, "b": 4, "c": 4}
print(all_equal(d))
```

```python
d = {"a": 4, "b": 6, "c": 4}
print(all_equal(d))
```

```python
d = {"a": 4, "b": 6, "c": 10}
print(all_equal(d))
```
