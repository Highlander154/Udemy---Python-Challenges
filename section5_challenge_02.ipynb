```python
from math import sqrt

def distance_ab(pointA: list[int], pointB: list[int]) -> float:
    dx = pointB[0] - pointA[0]
    dy = pointB[1] - pointA[1]
    dz = pointB[2] - pointA[2]
    distance = sqrt(dx**2 + dy**2 + dz**2)
    return round(distance, 5)
```
